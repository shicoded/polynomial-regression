```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate X values
# Create an array 'X' with 100 equally spaced values between -5 and 5
# np.linspace(start, stop, num) generates 'num' values from 'start' to 'stop'

X = np.linspace(-5, 5, 100)

# Step 2: Generate Y values
# Calculate 'Y' using a quadratic equation plus some random noise
# The equation Y = 2 * X^2 + 3 * X + 5 represents a quadratic relationship
# np.random.normal(mean, std, size) adds random noise to simulate real-world data

Y = 2 * X**2 + 3 * X + 5 + np.random.normal(0, 5, 100)

# Step 3: Plot the data
# Create a scatter plot of 'X' versus 'Y'
# 'plt.scatter' plots the data points

plt.scatter(X, Y, label='Data points')

# Add labels to the x-axis and y-axis
plt.xlabel('X')
plt.ylabel('Y')

# Add a title to the plot
plt.title('Synthetic Data with Quadratic Relationship')

# Add a legend to explain what the data points represent
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Step 1: Generate synthetic data
# Create an array 'X' with 100 equally spaced values between -5 and 5
X = np.linspace(-5, 5, 100)


# Calculate 'Y' using a quadratic equation plus some random noise
Y = 2 * X**2 + 3 * X + 5 + np.random.normal(0, 5, 100)

# Step 2: Transform input features to polynomial features
# PolynomialFeatures(degree) creates polynomial features of the specified degree
poly = PolynomialFeatures(degree=2)

# fit_transform reshapes 'X' and transforms it into polynomial features
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Step 3: Fit a linear regression model to the polynomial features
# Create an instance of LinearRegression
model = LinearRegression()
# Fit the model using the polynomial features 'X_poly' and the target values 'Y'
model.fit(X_poly, Y)

# Step 4: Predict Y values using the fitted model
# Predict 'Y_pred' using the fitted model
Y_pred = model.predict(X_poly)

# Step 5: Plot the synthetic data points and the fitted polynomial curve
# Create a scatter plot of 'X' versus 'Y'
plt.scatter(X, Y, label='Data points')
# Plot the polynomial regression curve using 'X' and 'Y_pred'
plt.plot(X, Y_pred, color='red', label='Fitted polynomial curve')
# Add labels and title to the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression Fit (Degree 2)')
plt.legend()
# Display the plot
plt.show()
```
